```python
class RateLimiter:
    def __init__(self, max_rate):
        self.semaphore = Semaphore(max_rate)

    def __enter__(self):
        self.semaphore.acquire()

    def __exit__(self, type, value, traceback):
        time.sleep(1)
        self.semaphore.release()

# Usage:
rate_limiter = RateLimiter(3)  
```

```python
printed_users = set()

# Get audio features
def fetch_song_features(track_id):
    audio_features_url = f'https://api.spotify.com/v1/audio-features/{track_id}'
    audio_features_response = requests.get(audio_features_url, headers=headers)
    return audio_features_response.json()

def populate_data(index_row):
    index, row = index_row
    if row['user_id'] not in printed_users:
        print("Current user index: ", my_split.index(row['user_id']))
        printed_users.add(row['user_id'])
    with rate_limiter:
        if not row['song_id']:
            return row
        api_response = fetch_song_features(row['song_id'])
        for feature in ['danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 'type', 'duration_ms', 'time_signature', 'analysis_url', 'uri', 'id']:
            if feature in api_response:
                row[feature] = api_response[feature]
        return row

with concurrent.futures.ThreadPoolExecutor() as executor:
    df = pd.DataFrame(list(executor.map(populate_data, df.iterrows())))
```
